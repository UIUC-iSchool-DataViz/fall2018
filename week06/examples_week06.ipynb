## Make a clock that says 2:10

```python
def make_clock(hours, minutes):
    arr1 = patches.Arrow(hours * (np.pi/6), 0, 0, 0.5)
    arr2 = patches.Arrow(minutes * (np.pi/30), 0, 0, 1.0)
    ax = plt.subplot(111,projection='polar')
    ax.set_theta_offset(np.pi/2)
    ax.set_theta_direction(-1)
    ax.add_patch(arr1)
    ax.add_patch(arr2)
    ax.set_xticks( [i * np.pi/6 for i in range(12) ]);
    ticklabels = [i for i in range(12)]
    ticklabels[0] = 12
    ax.set_xticklabels( ticklabels );
    ax.set_yticks( [] )
    ax.grid(False)
    return ax
```

```python
def make_stacked_chart(states, composition = True):
    base_line = np.zeros(duration_by_hour["il"].size)
    for state in states:
        plt.plot(base_line + duration_by_hour[state], label=state, drawstyle='steps')
        if composition:
            base_line += duration_by_hour[state]
    plt.yscale("log")
    plt.legend()
```

```python
make_stacked_chart(["il", "vt", "mi"])
```
